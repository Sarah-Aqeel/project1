```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/AB_NYC_2019 - AB_NYC_2019.csv')
df.head()
```

```python
# How many neighborhood groups are available and which shows up the most?
df['neighbourhood_group'].value_counts()

```

```python
# Are private rooms the most popular in manhattan?

df[df['neighbourhood_group']=='Manhattan']['room_type'].value_counts().idxmax()
```

```python
# Which hosts are the busiest and based on their reviews?
df.groupby('host_id')['number_of_reviews'].sum().reset_index().sort_values(by='number_of_reviews',ascending=False).head(10)
```

```python
df[df['host_id']==37312959]
```

```python
#Which neighorhood group has the highest average price?
df.groupby('neighbourhood_group')['price'].mean().idxmax()
```

```python
# Which neighborhood group has the highest total price?
df.groupby('neighbourhood_group')['price'].sum().idxmax()
```

```python
#Which top 5 hosts have the highest total price?
df.groupby('host_id')['price'].mean().reset_index().sort_values(by='price',ascending=False).head()
```

```python
df[df['host_id'].isin([72390391,20582832,5143901])]
```

```python
# Who currently has no (zero) availability with a review count of 100 or more?
df[(df['availability_365']==0) & (df['number_of_reviews']>=100)]

```

```python
# What host has the highest total of prices and where are they located?
df.groupby(['host_id','neighbourhood'])['price'].mean().reset_index().sort_values(by='price',ascending=False).head()
```

```python
# When did Danielle from Queens last receive a review?

df[(df['host_name']=='Danielle') & (df['neighbourhood_group']=='Queens')].sort_values(by='last_review',ascending=False)
#df[(df['availability_365']==0) & (df['number_of_reviews']>=100).value_counts()]
```
